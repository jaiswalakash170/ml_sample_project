### First Method

```python
fig = plt.figure() # Create a figure
ax = fig.add_subplot() # Add some axes
plt.show()
```

### Second Method

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x,y);
```

## Matplotlib example workflow

```python
# Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# Plot the data
ax.plot(x,y)

# Customize plot
ax.set(title="Simple Plot",
       xlabel="X-Axis",
       ylabel="Y-Axis")

# Save and show the whole figure
fig.savefig("./images/sample-plot.png")
```

### Two options for subplots

#### Subplot Option 1

```python
# Create figure with 4 different axes
fix, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))

# Plot to each different axes
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax4.hist(np.random.randn(1000));
```

#### Subplot Option 2

```python
# Create figure with 4 different axes
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))
# Plot to different indexes
ax[0,0].plot(x,x*2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

### Which one you should use pyplot ot matplotlib OO method?
* When plotting something quickly, okay to use pyplot method
* When plotting something more advanced, use OO method

```python
heart_disease["age"].plot.hist();
```

```python
heart_disease[:5]
```

```python
# Data of patients over age 50
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# Create scatter plot between age and chol - using pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target"); # c- Colour
```

```python
# Create a scatter plot between age and cholestrol - using OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
        x="age",
        y="chol",
        c="target",
        ax=ax)
# Change the limit of x-axis
ax.set_xlim([45,100]);
```

```python
# Create scatter plot between age and cholestrol using OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

#### Create a scatter plot between age vs cholestrol & thalach

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10))

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
        xlabel="Age",
        ylabel="CHolserol Levels");

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a meanline on ax0
ax0.axhline(y=over_50["chol"].mean(),
            linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline to ax1
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Subplot of chol, age, thalach with shared X-axis
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
        ylabel="CHolserol Levels");

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a meanline on ax0
ax0.axhline(y=over_50["chol"].mean(),
            linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline to ax1
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Reset the style
plt.style.use('seaborn-whitegrid')

# Create scatter plot between age and cholestrol using OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")# This changes the colour scheme
# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol") 

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
# Subplot of chol, age, thalach with shared X-axis
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                     cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
        ylabel="CHolserol Levels");

# Change the ax0 x axis limit
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target");

# Add a meanline on ax0
ax0.axhline(y=over_50["chol"].mean(),
            linestyle='--');

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")

# Change the ax1 x axis limit
ax1.set_xlim([50, 80])

# Change the ax1 y axis limit
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a meanline to ax1
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
